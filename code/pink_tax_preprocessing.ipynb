```python
import pandas as pd
import numpy as np
import os
# Work directory
os.chdir('../data/Datasets/Fragrances')
```

```python
df = pd.read_csv('data_pink_tax.csv', index_col=0)
```

```python
df
```

```python
# 1. Drop non relevant columns
df = df.drop(columns=['name', 'log_ml', 'new_price_per_ml', 'discount_per_ml'])

# Rename 'department' to 'gender' 
df = df.rename(columns={'department': 'gender'})

# Binary encoding for 'concentration' and 'gender'
df['concentration'] = df['concentration'].map({'EDP': 0, 'EDT': 1})
df['gender'] = df['gender'].map({'Men': 0, 'Women': 1})

# Map each brand to a unique number
df['brand'] = df['brand'].astype('category').cat.codes + 1  # Adds 1 to start numbering from 1

# Map each seller to a unique number
df['seller'] = df['seller'].astype('category').cat.codes + 1  # Adds 1 to start numbering from 1

# Map each scent to a unique number
df['scents'] = df['scents'].astype('category').cat.codes + 1  # Adds 1 to start numbering from 1

df.to_csv('data_pink_tax_preprocessed.csv', index=False) 

# Display the modified DataFrame
df.head()
```
